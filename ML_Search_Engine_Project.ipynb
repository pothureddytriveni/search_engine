```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])
    
    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
new_df['name'] = df['name'].str.replace(r'\([^)]*\)', '')  # Remove anything within parentheses
new_df['name'] = df['name'].str.replace(r'\.\d+', ' ')      # Remove any dot followed by numbers
new_df['name'] = df['name'].str.replace(r'[^\w\s]', ' ')    # Remove any non-alphanumeric characters except whitespace
```

```python
import re

def clean_text(text):
    if isinstance(text, bytes):
        # If the data is bytes-like, decode it assuming latin-1 encoding
        text = text.decode('latin-1')
        tet=text.lower()
    # Remove special symbols and extra spaces
    cleaned_text = re.sub(r'[^\w\s]', ' ', text)
    # Remove extra whitespace
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text.strip()
```

```python
# Clean 'file_content' column
new_df['cleaned_file_content'] = new_df['file_content'].apply(clean_text)

# Display the cleaned data
print(new_df['cleaned_file_content'])
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# Load your BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
model
```

```python
# Define a function to process each chunk
def process_chunk(chunk, model):
    # Generate embeddings for the current chunk
    chunk_embeddings = model.encode(chunk["cleaned_file_content"].tolist())
    # Extract document names from the chunk
    chunk_document_names = chunk["name"].tolist()
    return chunk_embeddings, chunk_document_names

# Define the size of each chunk
chunk_size = 500  # Adjust as needed based on memory constraints

# Split your dataset into chunks
chunks = [new_df[i:i + chunk_size] for i in range(0, len(new_df), chunk_size)]
```

```python
all_embeddings = []
all_document_names = []  # List to store document names
for chunk in chunks:
    try:
        chunk_embeddings, chunk_document_names = process_chunk(chunk,model)
        if chunk_embeddings is not None and len(chunk_embeddings) > 0:
            all_embeddings.append(chunk_embeddings)
            all_document_names.extend(chunk_document_names)  # Extend the list of document names
        else:
            print(f"No embeddings generated for chunk:\n{chunk}")
    except KeyError as e:
        print(f"Error processing chunk: {e}")
        print(f"Chunk size: {len(chunk)}")
        print(chunk)  # Print the chunk to inspect its contents
    except Exception as ex:
        print(f"An error occurred: {ex}")
```

```python
from chromadb import PersistentClient
from chromadb import chromadb

# Specify the directory where you want the ChromaDB database files to be stored
db_directory = r"C:\Users\lavak\Search Engine Project"

# Connect to ChromaDB
client = chromadb.PersistentClient(path=db_directory)

# Specify the name of the collection
collection_name = "document_embeddings"

# Create or get the collection
collection = client.get_or_create_collection(
    name=collection_name,
    metadata={"hnsw:space": "cosine"}  # Assuming cosine similarity
)

# Iterate through embeddings and document names
for doc_name, embeddings in zip(document_names, padded_embeddings ):
    try:
        # Add each document's embeddings to the collection
        for i, embedding in enumerate(embeddings):
            document_id = f"{doc_name}_{i}"  # Using a unique identifier for each embedding
            collection.add(document_id, embedding.tolist())
            print(f"Added embedding {i+1} for document: {doc_name}")
    except Exception as e:
        print(f"Error adding embeddings for document {doc_name}: {e}")
print("Embeddings added to the collection successfully.")
```
